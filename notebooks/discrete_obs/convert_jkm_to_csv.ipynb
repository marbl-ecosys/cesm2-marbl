```python
import pandas as pd
import numpy as np

from datetime import datetime, timezone
```

```python
filename = 'master_Fe_gx3v7.txt'

columns = ['month', 'lon', 'lat', 'depth', 'dFe_obs', 'gridx', 'gridy', 'gridlevel', 'inregion']
dtypes = {k: 'int32' for k in ['month', 'gridx', 'gridy', 'gridlevel', 'inregion']}

with open(filename) as fid:
    lines = []
    part = 0
    for line in fid:
        if part == 0:
            values = [float(s) for s in line.split()]
            part = 1
        else: 
            values.extend([float(s) for s in line.split()])
            values = np.array(values)
            values[values == -1e10] = np.nan
            lines.append({k: v for k, v in zip(columns, values)})
            part = 0

df = pd.DataFrame(lines).astype(dtypes).drop(['month', 'gridx', 'gridy', 'gridlevel', 'inregion'], axis=1)
df
```

```python
datestamp = datetime.now(timezone.utc).strftime("%Y-%m-%d")
df.to_csv(f'dFe-database-{datestamp}.csv', index=False, na_rep=-999.)
```
